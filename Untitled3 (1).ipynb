```python
def get_sentiment(text):
    prompt = f"Klasifikasikan sentimen ulasan ini. Jawab HANYA dengan satu kata: 'Positif' atau 'Negatif'.\nUlasan: '{text}'"
    try:
        output = replicate.run(
            "ibm-granite/granite-3.3-8b-instruct",
            input={"prompt": prompt, "max_new_tokens": 5}
        )
        return "".join(output).strip().replace(".", "")
    except Exception as e:
        return "Error"
```

```python
print("‚åõ Sedang melakukan klasifikasi sentimen... Mohon tunggu.")
df_subset = df.head(50).copy() # Anda bisa mengubah angka ini sesuai kebutuhan
df_subset['sentiment'] = df_subset['review_text'].apply(get_sentiment)
```

```python
plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")
ax = sns.countplot(data=df_subset, x='sentiment', palette="viridis")
plt.title("Distribusi Sentimen Ulasan Canon Powershot G3", fontsize=14)
plt.xlabel("Kategori Sentimen", fontsize=12)
plt.ylabel("Jumlah Ulasan", fontsize=12)
```

```python
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```
